```python
# Import statements
from selenium import webdriver
from selenium.webdriver import ActionChains
import pandas as pd
import time
```

```python
# Initialize the webdriver and specify browser
driver = webdriver.Chrome()
driver.get("https://fintechcanada.io/explore/?pg=1&country%5B%5D=Canada&sort=a-z")
```

```python
# Empty list to store company names
coNames = []

# counter var
j = 1

# While count is less than or equal to 47 (number of pages)
while (j <= 47):
    
    # Visit every page
    url = "https://fintechcanada.io/explore/?pg={}&country%5B%5D=Canada&sort=a-z".format(j)
    driver.get(url)
    
    # Find the company name
    coNames_element = driver.find_elements_by_xpath("//*[@class='case27-secondary-text listing-preview-title']")
    
    # Add it to the list of company names
    for i in range(len(coNames_element)):
        coNames.append(coNames_element[i].text)
    
    # Increment the counter
    j = j+1

# Print list of company names
print(coNames)
```

```python
# Empty list to store modified company names for url
co = []

# For each company name, change it to lower case (if has " &", delete it), replace space chars with "-"
for i in range(len(coNames)):
    co.append(coNames[i].lower().replace(" &", "").replace(" ", "-"))
    print(co[i])
```

```python
# Empty lists to store values of interest (industry sector, number of employees, incorporation year, city, and province)
indSect = []
numEm = []
incYr = []
city = []
prov = []

# For each company
for i in range(len(co)):
    
    # Visit each page
    url = "https://fintechcanada.io/listing/{}".format(co[i])
    driver.get(url)
    
    # Find the Industry sector
    indSect_element = driver.find_elements_by_xpath('/html/body/div[1]/div/div[3]/section[1]/div/div/div[1]/div/div[2]/div/div[2]/p[2]')

    # If it exists, add the text to indSect
    if indSect_element:
        indSect.append(indSect_element[0].text)
    else:
        indSect.append("")
    
    # Find the number of employees
    numEm_element = driver.find_elements_by_xpath('/html/body/div[1]/div/div[3]/section[1]/div/div/div[2]/div/div[2]/div/div[2]/p[2]')
    
    # If it exists, add the text to numEm
    if numEm_element:
        numEm.append(numEm_element[0].text)
    else:
        numEm.append("")
    
    # Find the incorporation year
    incYr_element = driver.find_elements_by_xpath('/html/body/div[1]/div/div[3]/section[1]/div/div/div[1]/div/div[3]/div/div[2]/p[2]')

    # If it exists, add the text to incYr
    if incYr_element:
        incYr.append(incYr_element[0].text)
    else:
        incYr.append("")
    
    # Find the city
    city_element = driver.find_elements_by_xpath('/html/body/div[1]/div/div[3]/section[1]/div/div/div[2]/div/div[1]/div/div[2]/ul/li[1]/div[2]/p')
    
    # If it exists, add the text to city
    if city_element:
        city.append(city_element[0].text)
    else:
        city.append("")
    
    # Find the province
    prov_element = driver.find_elements_by_xpath('/html/body/div[1]/div/div[3]/section[1]/div/div/div[2]/div/div[1]/div/div[2]/ul/li[2]/div[2]/p')
    
    # If it exists, add the text to prov
    if prov_element:
        prov.append(prov_element[0].text)
    else:
        prov.append("")

# Check length of each list
print(len(indSect))
print(len(incYr))
print(len(numEm))
print(len(city))
print(len(prov))
```

```python
# Convert the information collected to a data frame
df = pd.DataFrame(list(zip(coNames, city, prov, incYr, numEm, indSect)), 
                  columns=["Company", "City", "Province", "Incorporation Year", "Number of Employees", "Industry Sector"])

# Display
df
```

```python
# Write to csv
df.to_csv(r'FintechInfo.csv', index=False)
```

```python
# Empty lists to store information
comp = []
bn = []
regID = []
ct = []
p = []
status = []
statusNotes = []
bizType = []
created = []

# For each company
for i in range(len(coNames)):
    
    # Visit each page
    url = "https://beta.canadasbusinessregistries.ca/search/results?search=%7B{}%7D&status=All".format(coNames[i])
    driver.get(url)
    
    # Wait 1 second
    time.sleep(1)
    
    # Click the "I accept the term of use button"
    button = driver.find_elements_by_xpath('//*[contains(text(), "I accept the terms of use")]')
    
    # If the button pops up
    if button:
        button[0].click()
    
    # Wait 3 seconds
    time.sleep(1)
    
    # Find the Company name (In case it matches with the wrong one)
    comp_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/div/span')
    
    # If it exists, add the text to comp
    if comp_element:
        comp.append(comp_element[0].text)
    else:
        comp.append("")
    
    # Find the Business Number
    bn_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[1]/span')

    # If it exists, add the text to bn
    if bn_element:
        bn.append(bn_element[0].text)
    else:
        bn.append("")
    
    # Find the registry ID
    regID_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[2]/span')
    
    # If it exists, add the text to regID
    if regID_element:
        regID.append(regID_element[0].text)
    else:
        regID.append("")
    
    # Find the location
    ct_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[3]/span[1]')

    # If it exists, split it by city and province and add it to respective list
    if ct_element:
        ct.append(ct_element[0].text.replace(", ", ""))
    else:
        ct.append("")
    
    # Find the province
    p_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[3]/span[2]')
    
    # If it exists, add the text to prov
    if p_element:
        p.append(p_element[0].text)
    else:
        p.append("")
    
    # Find the Status
    status_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[4]/span')
    
    # If it exists, add the text to status
    if status_element:
        status.append(status_element[0].text)
    else:
        status.append("")
        
    # Find the Status Notes
    statusNotes_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[5]/span')
    
    # If it exists, add the text to statusNotes
    if statusNotes_element:
        statusNotes.append(statusNotes_element[0].text)
    else:
        statusNotes.append("")
    
    # Find the Business Type
    bizType_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[6]/span')
    
    # If it exists, add the text to bizType
    if bizType_element:
        bizType.append(bizType_element[0].text)
    else:
        bizType.append("")

    # Find the Created date
    created_element = driver.find_elements_by_xpath('/html/body/mras-root/main/mras-search-results/div/div/mras-primary-card/article/div/div/ul/li[7]/span')
    
    # If it exists, add the text to bizType
    if created_element:
        created.append(created_element[0].text)
    else:
        created.append("")
```

```python
df3 = pd.DataFrame(list(zip(coNames, comp, bn, regID, ct, p, status, statusNotes, bizType, created)), 
                  columns=["Company", "Matched Company Name", "BN", "regID", "City", "Province", "Status", 
                           "Status Notes", "Business Type", "Created"])

df3
```

```python
df3.to_csv(r'FinTechBN.csv', index=False)
```
